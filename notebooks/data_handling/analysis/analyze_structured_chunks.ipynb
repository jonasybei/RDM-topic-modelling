```python
import pandas as pd
from git_root import git_root
import matplotlib.pyplot as plt
import nltk
import numpy as np
```

```python
my_git_root = git_root()
```

```python
documents_chunked_path = f'{my_git_root}/data/documents_chunked.csv'
documents_chunked = pd.read_csv(documents_chunked_path)
```

```python
def get_num_sentences(document):
    return len(nltk.sent_tokenize(document))
```

```python
chunks = documents_chunked['text']
num_tokens = documents_chunked['num_tokens'].to_list()
num_sentences = [get_num_sentences(chunk) for chunk in chunks]
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

ax[0].hist(num_tokens, bins=np.arange(0, np.max(num_tokens) + 50, 50))
ax[0].set_xticks(np.arange(0, np.max(num_tokens) + 50, 50))
ax[0].grid()

ax[1].hist(num_sentences, bins=np.arange(0, np.max(num_sentences) + 2, 2))
ax[1].set_xticks(np.arange(0, np.max(num_sentences) + 2, 2))
ax[1].grid()
plt.savefig(f'{my_git_root}/notebooks/images/chunking_statistics/structured_chunking_num_tokens.svg')
plt.show()
```
