```python
import pymupdf4llm
import os
import glob
from git_root import git_root
```

```python
my_git_root = git_root()
```

```python
def get_filenames(location, filetype):
    filenames = set()
    os.chdir(location)
    for filename in glob.glob(f'*.{filetype}'):
        filenames.add(filename)
    return sorted(filenames)
```

```python
pdf_location = f'{my_git_root}/data/pdf/'
filenames = get_filenames(pdf_location, 'pdf')
```

```python
documents = []

for filename in filenames:
    text = pymupdf4llm.to_markdown(pdf_location + filename)
    documents.append(text)
```

```python
import pandas as pd

df = pd.DataFrame(data={
    'name': filenames,
    'text': documents
})
```
